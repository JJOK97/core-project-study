```python
#res.text는 String 데이터
#★★★★★bs를 이용해서 String -> html로 변환하여 접근 가능!
from bs4 import BeautifulSoup as bs
```

```python
allNews = req.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9D%B4%EC%83%81%EA%B8%B0%ED%9B%84")
```

```python
#bs(변환할 데이터, 파싱 방법)
#파싱(parsing) : 데이터를 특정 규칙에 맞게 구조화된 형태로 바꾸는 작업
soup = bs(allNews.text, "lxml")
```

```python
# 파싱 결과 : 객체 {태그:<span>, 속성:{align:center}, 콘텐츠:"~~"}
# -> 컴퓨터가 각각의 키값을 통해 값에 접근할 수 있음.

#파싱을 하는 이유 : 컴퓨터에게 각각 태그, 속성, 컨텐츠에 접근을 시키기 위해서
soup
```

```python
# select("선택자")
# HTML데이터에서 특정 요소(태그, 콘텐츠)만 수집
# 명확하게 요청하지 않으면 정확한 수집이 불가능

news_title = soup.select("a.news_tit")  # a태그의 news_tit 클래스
```

```python
news_title
```

```python
len(news_title)
```

```python
# 우리가 수집한 news_title 안에는 "요소"가 들어있음!
# 우리가 필요한 건? content
# 해결법!! 수집된 "요소"에서 "content"만 뽑아내자!

news_title[0].text
```

```python
#반복문을 활용해서 모든 제목 데이터만 출력
# +활용을 위해 제목 텍스트만 담을 리스트를생성 

title_list = []

for i in news_title :
    print(i.text)
    title_list.append(i.text)
    
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(title_list,column=["제목"])
```
