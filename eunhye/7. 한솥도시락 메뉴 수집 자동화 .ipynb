```python
# 1. 필요한 라이브러리 import

!pip install selenium
from selenium import webdriver as wb
# 키보드 라이브러리"
from selenium.webdriver.common.keys import Keys
# 선택자 라이브러리
from selenium.webdriver.common.by import By

import time
```

```python
# 2. 크롬 브라우저 실행
#크롬 창에 대한 역할
driver = wb.Chrome()
```

```python
# 3. "한솥 메뉴 사이트 주소를 주고 이동하라고 시킴
driver.get("https://www.hsd.co.kr/menu/menu_list")
# #btn_more > span > a
```

```python
# 더보기 탭 클릭해보기 
# 특정 요소를 선택할때, 개발자 도구에서 태그 우클릭 > [copy] -> [copy selector] 누르면
# 그 요소의 고유 선택자를 가져올 수 있는데, 복사한 전체를 사용해야 할지, 아니면 뒷부분만 사용해도 될지는 판단해보기!
# #btn_more > span > a
tab = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")
tab.click()
```

```python
# 4. '더보기' 버튼 다섯 번 클릭(쉬어가면서~)

body = driver.find_element(By.CSS_SELECTOR, "#btn_more > span > a")

for i in range(5):
    body.click()
    time.sleep(1)
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
res = req.get("https://www.hsd.co.kr/menu/menu_list")
```

```python
soup = bs(res.text, "lxml")
```

```python
soup
```

```python
items = soup.select("li[id^='child_'] a")
```

```python
item_list = []
```

```python
for item in items:
    print(item.text.strip())

for item in items :
    item_list.append(item.text)

item_list
```

```python
price_list = []
```

```python
menu_prices = driver.find_elements(By.CSS_SELECTOR, "div.item-text > div > strong")
```

```python
# 출력
for menu_price in menu_prices:
    print(menu_price.text.strip())

for menu_price in menu_prices :
    price_list.append(menu_price.text)

price_list
```
