#### 틱(tick) 설정
- 플롯이나 차트에서 축 상의 위치 표시 지점 : 틱
- 자동으로 설정, 수동 설정 가능

```python
x = np.arange(7)
y = [1, 4, 7, 8, 9, 5,23]
plt.plot(x, y, markersize=5, markerfacecolor='yellow', markeredgecolor='yellow', markeredgewidth=3, linewidth=1, linestyle='--')
plt.xticks([0,3,6])
plt.yticks([1 ,5 ,20])
plt.show()
```

```python
# x, y 축 그래프 제목 설정
x = np.arange(7)
y = [1, 4, 7, 8, 9, 5,23]
plt.plot(x, y, markersize=5, markerfacecolor='yellow', markeredgecolor='yellow', markeredgewidth=3, linewidth=1, linestyle='--')
plt.xticks([0,3,6])
plt.yticks([1 ,5 ,20])

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X-Y 가나다')

plt.show()
```

### 여러개 플롯 한꺼번에 그리기

```python
x = np.arange(4)
y1 = [1, 6, 3, 12]
y2 = [5, 2, 11, 8]
y3 = [9, 10, 7, 4]

plt.plot(x, y1, label = 'y1', color='red')
plt.plot(x, y2, label = 'y2', color='blue')
plt.plot(x, y3, label = 'y3', color='green')

# 범례 설정 legend
plt.legend(loc='upper center', prop = {'size':10})
# 범례의 위치 : loc


plt.show()
```

#### 바 차트(bar chart) : 막대그래프

```python
x = np.arange(6)
y = [80 ,90, 35, 60, 85, 50]

# bar chart 만들어주는 함수}
plt.bar(x, y, color=['green', 'orange', 'purple'], edgecolor='blue', linewidth = 3, alpha = 0.5, width = 1.3)

# color: 바 내부 색상
# edgecolor : 테두리 색상
# linewidth : 테두리 너비
# alpha : 투명도( 0 ~ 1, 기본값 1 )
# width : 너비
```

```python
car_car = data.loc[data['사고유형_대분류'] == '차대차']
car_car
```

```python
# 데이터 프레임 모든 열(컬럼) (일부열 가능) 출력
pd.set_option('display.max_columns', None)
```

```python
car_daytime = car_car.loc[car_car['주야'] == '주간']

car_daytime
```

```python
car_daytime.value_counts('요일')
```

```python
# 요일 순서를 정렬
date = car_daytime['요일'].value_counts()

date
```

```python
date = date[['월', '화', '수', '목','금','토','일']]
```

```python
from matplotlib import rc

rc('font', family = 'Malgun Gothic')
```

```python
y = date.values
x = date.index

plt.bar(x, y)

plt.ylim(0, 180)
plt.title('요일별 주간 차대차 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')

plt.show()
```
