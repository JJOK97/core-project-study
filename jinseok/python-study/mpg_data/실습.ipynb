#### 5개 제조사별로 생산하는 자동차 종류를 시각화


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sns.countplot(data = , x = , hue = )
# x : x 값
# hue : 분할 기준

sns.countplot(data = mpgSetIndex, x = 'manufacturer', hue = 'category')
# x 축 : 제조사
# 분할 기준 : 자동차 종류 ( 종류별로 분할해서 개수를 나타내라 )
```

```python
# 만약 데이터가 결측치가 있을 경우 참고하기 좋다.
# 결측치는 일반적으로 평균, 중앙값, 최빈값을 넣을 수 있지만,
# 시각화 데이터 분석을 참고하여 채울 수 있다.

# 자동차 종류에 결측치가 있다고 가정,
# 1. 제조사가 audi -> compact일 확률이 높다
# 2. 제조사가 honda일 경우, subcompact일 확률이 높다.
```

#### 제조사별, 자동차 종류별, 고속도로 연비 시각화
- violinplot으로 확인해보자
- violinplot : 수치형 데이터의 '분포'를 확인할 때 사용하는 그래프

```python
# 바이올린 형태로 데이터의 분포와 밀도를 곡선으로 표현
# 데이터가 많이 분포 -> 두껍다
# 데이터가 적게 분포 -> 얇다

# 제조사와 자동차 종류에 따른 고속도로 연비

plt.figure(figsize = (15, 10))

sns.violinplot(
    data=mpgSetIndex,
    x = 'manufacturer',
    y = 'cty',
    hue = 'category', # 분류 기준( 나타내고자 하는 것 )
    inner = 'stick', # 바이올린 내부에 개별 데이터 선
)
```
