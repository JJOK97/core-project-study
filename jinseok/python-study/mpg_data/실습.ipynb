```python
# q4. 회사별 compact 차종 수를 내림차순

# groupby('제조사별').mean()
# groupby('제조사별').count()

'''
1. category == 'compact'
2. manufacturer, category 데이터
3. groupby('제조사별').count
4. category(개수)별로 내림차순
'''

mpg[['manufacturer', 'category']].groupby('manufacturer').count()
```
