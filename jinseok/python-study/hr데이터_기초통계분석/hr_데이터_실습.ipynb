#### 1. 데이터 병합
- concat
- 4, 5, 6월 pay-info 데이터를 각각 불러와 한개의 df 병합

```python
file_names = os.listdir('월별지급')
file_names
```

```python
pay_df_list = []
for fn in file_names:
    df = pd.read_csv('월별지급/' + fn, encoding = 'euc-kr')
    pay_df_list.append(df)
```

```python
pay_df_list
```

```python
# 데이터 병합
total_df = pd.concat(pay_df_list)

total_df
```

```python
# csv 파일로 저장
total_df.to_csv('total_pay_info.csv')
```

```python
#### org_info 데이터 (조직정보)에 각 조직별 조직 인원 수, 평균 나이, 평균 근무 년수, 평균 연봉
```

```python
# 조직 정보 데이터 확인
org_info = pd.read_csv('org-info.csv', encoding = 'euc-kr', index_col = 'ORG_CD')
org_info
```

```python
hr_info = pd.read_csv('hr-info.csv', encoding = 'euc-kr', index_col = 'ORG_CD')
hr_info
```

```python
hr_info.groupby('ORG_CD').count()
```

```python
# 조직 정보 데이터와 공통적으로 갖고 있는 컬럼 : ORG_CD
# ORG_CD 컬럼을 기준으로 groupby

# hr_info에서 groupby를 이용해서 조직별 조직 인원 수, 평균 나이, 평균 근무 년수, 평균 연봉 추가
org_agg = hr_info.groupby('ORG_CD').agg({'ENAME' : 'count',
                                         'AGE' : 'mean',
                                        'SERVICE_YEAR' : 'mean',
                                        'SALARY' : 'mean'})

org_agg

```

```python
org_info.head(1)
```

```python
org_info
```

```python
# org_info와 조직구성원 데이터서에서 추출한 df를 열방향 병합
# 주의 : concat을 이용한 열 방향 병합 시, 각 DataFrame에 인덱스를 같게 해야 올바른 병합이 일어남

org_info_new = pd.concat([org_info, org_agg], axis = 1)
org_info_new
```

```python
# 급여 지급내역 데이터 로딩
hr_pay_info = pd.read_csv('hr-pay-info.csv', encoding='euc-kr')

hr_pay_info
```

```python
hr_info
```

```python
set(hr_info.columns) - (set(hr_info.columns) - set(hr_pay_info.columns))
```

```python
# 둘 다 가지고 있는 EID 컬럼을 중심으로 merge
# 사원 정보데이터에서, 사용할 EID, 이름, 성별
hr_info_df = hr_info[['EID','ENAME','GENDER']]
hr_info_df
```

```python
# 사원 정보데이터 기준으로 merge
total_salary_df1 = pd.merge(hr_info_df, hr_pay_info, how='left', on='EID')
total_salary_df1

# 병합 시, 'EID' 열을 기준으로 병합, 왼쪽 데이터 기준
```

```python
# 급여지급 내역을 중심으로 merge
total_salary_df2 = pd.merge(hr_info_df, hr_pay_info, how='right', on='EID')
total_salary_df2
```

```python
# 급여 지금 내역을 중심으로 merge 했을 때 더 적음
# -> 급여지급이 안된 인원들이 있다.
```

```python
# 지급내역이 없는 인원들을 확인
total_salary_df1['PAYMENT_DY'].isnull()

# null인지 확인
```

```python
# 불리언 인덱싱
no_money = total_salary_df1[total_salary_df1['PAYMENT_DY'].isnull()]
no_money
```

```python
# 지급 내역이 없는 사람들이 언제 퇴사를 했는지 알아보자

pd.merge(no_money, hr_info)[['ENAME', 'RETIRE_DATE']]
# 퇴사자여서 급여지급이 없었다.
```
