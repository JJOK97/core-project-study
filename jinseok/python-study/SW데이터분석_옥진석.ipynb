```python
# 문제 3
list1 = [1, 50, 27, 19, 37, 33, 5, 3, 48]

def countOddAndEven(list):
    odd = 0
    even = 0
    for i in list:
        if i % 2 == 0:
            even += 1
        elif i % 2 != 0:
            odd += 1

    print(f"짝수 {even}개, 홀수 {odd}개")

countOddAndEven(list1)
```

```python
# 문제 4
score_list = [['김미희', 89, 78, 99],
              ['서대희', 67, 88, 77],
              ['강예진', 100, 55, 96],
              ['차현석', 75, 100, 60]]


def calScore(score_list):

    score_list.insert(0, ['이름', 'Python', 'Java', 'DB', '평균', '학점'])

    python_sum = 0
    java_sum = 0
    db_sum = 0

    for i in range(1, len(score_list)):
        student = score_list[i]

        python_sum += student[1]
        java_sum += student[2]
        db_sum += student[3]

        avg = int((student[1] + student[2] + student[3]) / 3)

        if avg >= 90:
            grade = 'A'
        elif avg >= 80:
            grade = 'B'
        elif avg >= 70:
            grade = 'C'
        elif avg >= 60:
            grade = 'D'
        else:
            grade = 'F'

        student.append(avg)
        student.append(grade)

    num_students = len(score_list) - 1

    python_avg = int(python_sum / num_students)
    java_avg = int(java_sum / num_students)
    db_avg = int(db_sum / num_students)

    avg_row = ['평균', python_avg, java_avg, db_avg, '', '']
    score_list.append(avg_row)

    return score_list


table = calScore(score_list)

print('=====================================================')
print("%-8s%-8s%-8s%-8s%-8s%-8s" % ("이름", "Python", "Java", "DB", "평균", "학점"))
print('=====================================================')

for t in range(1, len(table)-1):
    print("%-8s%-8s%-8s%-8s%-10s%-8s" % (score_list[t][0], score_list[t][1],score_list[t][2], score_list[t][3], score_list[t][4], score_list[t][5]))

print('=====================================================')
print("%-8s%-8s%-8s%-8s" % ("평균", score_list[len(score_list)-1][1], score_list[len(score_list)-1][2], score_list[len(score_list)-1][3]))
```

```python
# 문제 5
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

X = [2, 3, 4, 5, 6, 7, 8]
Y1 = [5, 7, 3, 9, 1, 8, 6]
Y2 = [9, 5, 7, 2, 3, 7, 1]

plt.plot(X, Y1, label = 'Y1', color='red')
plt.plot(X, Y2, label = 'Y2', color='blue')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid()
plt.legend(loc='upper right', prop = {'size':10})

plt.show()
```

```python
# 문제 6
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')

data = pd.read_csv('C:\\Users\\1\\Desktop\\workspace\\core-project-study\\jinseok\\python-study\\data\\Traffic_Accident_2017.csv', encoding = 'euc-kr')

car_human = data.loc[data['사고유형_대분류'] == '차대사람']

car_night_time = car_human.loc[car_human['주야'] == '야간']

date = car_night_time.value_counts('요일')

date = date[['월', '화', '수', '목','금','토','일']]

y = date.values
x = date.index

plt.bar(x, y)

plt.ylim(0, 200)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')

plt.show()
```
